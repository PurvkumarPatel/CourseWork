```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.subplot(1,2,1)
plt.scatter(x, y, c=colors, cmap='nipy_spectral',s=sizes,alpha=0.7)
plt.colorbar()
plt.subplot(1,2,2)
plt.scatter(x, y, c=colors, cmap='tab10_r',s=sizes,alpha=0.3)

plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,'a',103,87,94,78,77,85,86])
plt.subplot(1,2,1)
plt.bar(x, y, color ='r',width=0.5)
plt.subplot(1,2,2)
plt.barh(x, y, color ='b',height=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x,color='m')
plt.ylabel('Height')
plt.xlabel('Persons')
plt.title('Average Height in Histogram')
plt.grid(axis='x',ls=':')
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.1, 0, 0.5]
mycolor = ['r','yellow','r','brown']

plt.pie(y, labels = mylabels, explode = myexplode, shadow = 1, colors = mycolor)
plt.legend(title = 'Fruits')
plt.show() 
```
